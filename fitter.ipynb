```python
from scipy import stats
from scipy.optimize import curve_fit
from scipy.stats import norm       

max_t = data['\epsilon'].copy()
gauss_range = [-1, 1]

fig = plt.figure(figsize=(12,7))
weights = np.ones_like(max_t)/float(len(max_t))
binning = np.linspace(-10, 10)
n, bins, patches = plt.hist(max_t, 200, alpha=0.75, weights=weights)
step = bins[1] - bins[0]
x = (0.5*(bins[1:]+bins[:-1]))
mask = np.logical_and(x >= gauss_range[0], x <= gauss_range[1])
y = n[mask]
x = x[mask]
y = y[np.abs(x) > 0.05]
x = x[np.abs(x) > 0.05]

x1 = x.copy()
y1 = y.copy()

mean = sum(x * y) / sum(y)
sigma = np.sqrt(sum(y * (x - mean)**2) / sum(y))

def Gauss(x, a, x0, sigma):
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))

popt,pcov = curve_fit(Gauss, x, y, p0=[max(y), mean, sigma])
plt.plot([], [], linestyle='none', marker=None, label="Filter: %s\nPoints: $%s$\nGauss fit: $A=%.2f$, $x_0=%.2f$, $\sigma=%.2f$" % (filter_str, len(max_t), popt[0], popt[1], popt[2]))
plt.legend(frameon=False, loc='upper left')
_x = np.linspace(-1, 1, 1000)
plt.plot(_x, Gauss(_x, *popt), color='orange')
plt.plot()
plt.xlabel('$\epsilon$')
plt.ylabel('probability')
plt.xlim([-0.25, -0.05])
#plt.yscale('log')
plt.ylim([None, 0.02])
plt.show()
```

```python
R0 = 1
y = 1
a = 2.1

def f1():
    ret = []
    for y in range(1, 1000):
        f = lambda r: np.abs(r * (1 - np.abs(R0/r) ** a) - y)
        r1 = optimize.fmin(f, -R0, full_output=True, xtol=1e-9, ftol=1e-9, maxiter=10000, maxfun=10000, disp=False)
        r2 = optimize.fmin(f, R0 + 1e-10, full_output=True, xtol=1e-9, ftol=1e-9, maxiter=10000, maxfun=10000, disp=False)
        r1 = r1[0][0]
        r2 = r2[0][0]
        ret.append([y, r1, r2])
#    print("Solution:", r1, r2)
#    print("Result:", f(r1), f(r2))
    return np.array(ret)

def f2():
    ret = []
    for y in range(1, 1000):
        f = lambda r: np.abs(np.abs(r * (1 - np.abs(R0/r) ** a)) - y)
        r1 = optimize.fminbound(f, 1e-12, R0, full_output=True,xtol=1e-9,maxfun=10000,disp=False)
        r2 = optimize.fminbound(f, R0, 1e12, full_output=True,xtol=1e-9,maxfun=10000,disp=False)
        r1 = r1[0]
        r2 = r2[0]
        ret.append([y, r1, r2])
#    print("Solution:", r1, r2)
#    print("Result:", f(r1), f(r2))
    return np.array(ret)
```
